```python
#plt plot
plt.plot(x,y,'ro',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
#Creatig subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'go')
plt.subplot(2,2,3)
plt.plot(x,y,'b*')
plt.subplot(2,2,4)
plt.plot(x,y,'bo--')
```

```python
#Compute x and y cordinates for the ponts on a sine curve
x= np.arange(0,3 * np.pi,0.1)
y= np.sin(x)
plt.title("sine wave form")
# plotted using matplolib 
plt.plot(x,y)
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
```

```python
## Bar graph

x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y,color= 'r' ) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis') 
```

# Histogram

```python
data = [np.random.normal(0, std, 100) for std in range(1,5)]

plt.boxplot(data,vert=True)
```

```python
# Data to plot
labels = 'Python', 'C++', 'JavaScript ', 'Java'
sizes = [21, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.9, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
